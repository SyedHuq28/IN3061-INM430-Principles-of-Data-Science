```python
import pandas as pd
df1 = pd.read_csv('passengerData.csv')
```

```python
df2 = pd.read_excel('ticketPrices.xlsx')
df2
```

```python
new = pd.merge(df1, df2, on='TicketType')
```

```python
import matplotlib.pyplot as plt

ax = plt.axes
plt.scatter(new['Fare'], new.Age)
plt.xlabel("Fare")
plt.ylabel("Age")
plt.show()
```

```python
female_new  = new.loc[(new.Age > 40) & (new.Age < 50) & (new.Sex == 'female') & (new.Fare >= 40)]
female_new
```
