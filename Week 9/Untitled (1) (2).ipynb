```python
import matplotlib.pyplot as plt
G = nx.petersen_graph()
subax1 = plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.clear()
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.add_node("A") 
G.add_node("B") 
G.add_node("C") 
G.add_node("D") 
G.add_node("E") 
G.add_node("F") 
G.add_node("G") 
G.add_node("H") 
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.add_weighted_edges_from([("A","B",1),("A","C",3),("C","B",7),("C","E",4),("C","D",9),("D","F",8),("F","G",2),("F","H",5)])
nx.draw(G, with_labels=True)
```

```python
nx.degree_centrality(G) 
```

```python
fin_max = max(nx.degree_centrality(G), key=nx.degree_centrality(G).get)
print("Maximum centrality:",fin_max)
```

```python
mb = nx.betweenness_centrality(G, k=None, normalized=True, weight=None, endpoints=False, seed=None)
mb
```

```python
fin_max = max(mb, key=mb.get)
print("Maximum betweenness centrality:",fin_max)
```
