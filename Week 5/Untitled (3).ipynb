```python
import matplotlib.pyplot as plt
plt.scatter(df.GrLivArea, df.SalePrice)
```

```python
df['isOutlierGrLivArea'] = 0
df['isOutlierSalePrice'] = 0
```

```python
std1 = df['GrLivArea'].std()
mean1 = df['GrLivArea'].mean() 
st1 = mean1 - 3*std1
st2 = mean1 + 3*std1
df.loc[(df.GrLivArea < st1) & (df.GrLivArea > st2) , 'isOutlierGrLivArea' ] = 1
```

```python
std2 = df['SalePrice'].std()
mean2 = df['SalePrice'].mean() 
st3 = mean2 - 3*std2
st4 = mean2 + 3*std2
df.loc[(df.SalePrice < st3) & (df.SalePrice > st4) , 'isOutlierSalePrice' ] = 1
```

```python
df2 = df[df.isOutlierGrLivArea == 0]
```

```python
df2 = df[(df.isOutlierGrLivArea == 0)  & (df.isOutlierSalePrice == 0)]
```
